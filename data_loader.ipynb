```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# Set up headless Chrome
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
driver = webdriver.Chrome(options=options)

# Load the page
url = "https://technofies.com/blog/digital-marketing-experts-in-pakistan/"
driver.get(url)

# Wait for JavaScript to load content
time.sleep(5)  # Adjust if needed

# Get fully rendered HTML
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Extract meaningful text
important_text = []

title = soup.title.string if soup.title else ""
important_text.append(title)

for tag in soup.find_all(['h1', 'h2', 'h3', 'p']):
    text = tag.get_text(strip=True)
    if text and len(text.split()) > 5:
        important_text.append(text)

driver.quit()

final_content = "\n\n".join(important_text)
print(final_content)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
import time
import os
```

```python
os.environ["HF_TOKEN"] = os.getenv('HF_TOKEN')
```

```python
urls = [
    "https://technofies.com/blog/digital-marketing-experts-in-pakistan/",
    "https://thedesignsfirm.com/scope-of-digital-marketing-in-pakistan/",
    "https://www.digitalmediatrend.com/why-do-small-businesses-struggle-with-digital-marketing-in-pakistan/",
    "https://www.statista.com/outlook/amo/advertising/social-media-advertising/pakistan",
    "https://brandbuilders.com.pk/case-study-of-successful-online-marketing-campaigns-in-pakistan/",
    "https://brandbuilders.com.pk/how-a-digital-marketing-agency-can-boost-sales-in-pakistan/"
]
```

```python

options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

docs = []

# Step 3: Scrape and extract meaningful content
for url in urls:
    driver.get(url)
    time.sleep(5)  # wait for JS
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Title
    title = soup.title.string.strip() if soup.title else "No title"

    # Headings and paragraphs
    content = [title]
    for tag in soup.find_all(['h1', 'h2', 'h3', 'p']):
        text = tag.get_text(strip=True)
        if text and len(text.split()) > 5:
            content.append(text)

    full_text = "\n".join(content)

    # Create a LangChain document with metadata
    doc = Document(page_content=full_text, metadata={"source": url})
    docs.append(doc)

driver.quit()


```

```python
from pathlib import Path
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunks = splitter.split_documents(docs)
```

```python
# Step 4: Embed and store in FAISS
from langchain.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
# Optional: Save to disk
vector_store.save_local("faiss_digital_marketing_knowledgebase")
```
